```python
import requests
from bs4 import BeautifulSoup

URL = "https://arxiv.org/list/cs.AI/recent"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

results = soup.find(id='articles')
print(results.prettify())
```

```python
links=soup.find_all("a", string="pdf")
links_list=[]
for element in links:
    links_list.append('https://arxiv.org/'+element['href'])
links_list
```

```python

list = results.find_all("div", class_="list-authors")
articles=[]
for item in list:
    authors=[]
    for element in item:
        if element.text!=", ":
            authors.append(element.text)
    articles.append(authors)
articles
```

```python

title = results.find_all("div", class_="list-title mathjax")
titles_list=[]
for item in title:
    titles_list.append(item.text.replace('Title:\n          ', '').replace('\n        ', '').strip())

titles_list
```
